```python
data = pd.read_csv('1256752196.csv')

data['retail_greater_than_contr'] = (data['retail_price'] - data['contr_amount']) > 0 # 261 of times retail is higher then contract

data['retail_greater_than_contr'].value_counts()

data.head(100)[['retail_price', 'contr_amount', 'retail_greater_than_contr']]
```

## Missing values

```python
# Count missing values of retail_price #125
print(f"% Missing values for retail price: {data['retail_price'].isna().sum()*100 / data.shape[0]}%")
print(f"% Missing values for contract amount: {data['contr_amount'].isna().sum()*100 / data.shape[0]}%")
```
