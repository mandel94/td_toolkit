```python
import os
import sys
import pandas as pd


# Add two levels above to path 
sys.path.append(os.path.abspath(os.path.join(os.path.dirname('.'), '..', '..')))
```

```python
from config import WEEKLY_OUTPUT_DIR, WEEKLY_REPORT_OUTPUT_FILENAME

output_path = os.path.join(WEEKLY_OUTPUT_DIR, WEEKLY_REPORT_OUTPUT_FILENAME)

weekly_data = pd.read_csv(output_path)

weekly_data.Categoria.value_counts()
```

```python
category = "Recensioni / In Sala"

# Filter
filtered_weekly_data = weekly_data[weekly_data["Categoria"] == category]
if category == "News":
    filtered_weekly_data = filtered_weekly_data[~filtered_weekly_data["Percorso pagina e classe schermata"].str.contains("si-fara")]
for path in filtered_weekly_data.nlargest(columns="Utenti attivi", n=20)["Percorso pagina e classe schermata"]:
    print(path)
```
