```python
import pandas as pd


may24_to_may25_data = pd.read_csv(data_path + '\\may24_to_may25_articles.csv')

may24_to_may25_data
```

```python
# Convert pubdate to datetime
may24_to_may25_data['pubdate'] = pd.to_datetime(may24_to_may25_data['pubdate'])

may24_to_may25_data
```

## Analisi orario di pubblicazione articoli

```python
may24_to_may25_data["time_of_day"] = may24_to_may25_data["pubdate"].dt.hour
# Plot histogram of time of day
may24_to_may25_data["time_of_day"].hist(bins=24, edgecolor='black')
import matplotlib.pyplot as plt
plt.xlabel('Hour of Day')
plt.ylabel('Number of Articles')
plt.title('Distribution of Articles by Hour of Day')
plt.xticks(range(0, 24))
plt.grid(axis='y')

plt.show()
```

```python
active_users_by_tod = may24_to_may25_data.groupby('time_of_day')['active users'].agg(['mean', 'median', 'std', 'count'])
```

```python
# Plot combo chart, with mean and median as lines, max value as shaded area, and count as bars
plt.figure(figsize=(12, 6))
plt.bar(active_users_by_tod.index, active_users_by_tod['count'], color='lightgray', label='Count', alpha=0.5)
plt.plot(active_users_by_tod.index, active_users_by_tod['mean'], color='blue', label='Mean', linewidth=2)
plt.plot(active_users_by_tod.index, active_users_by_tod['median'], color='orange', label='Median', linewidth=2)

plt.xlabel('Hour of Day')
plt.ylabel('Active Users')
plt.title('Active Users by Hour of Day')
plt.xticks(range(0, 24))
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Plot boxplot of active users by hour of day
plt.figure(figsize=(12, 6))
may24_to_may25_data.boxplot(column='active users', by='time_of_day', grid=False)
plt.xlabel('Hour of Day')
plt.ylabel('Active Users')
plt.title('Boxplot of Active Users by Hour of Day')
plt.suptitle('')
plt.xticks(range(0, 24))
plt.grid(axis='y')
plt.show()
```

```python
# Test is significant difference between active users and time of day
import numpy as np
from scipy.stats import f_oneway
mask = ~np.isnan(may24_to_may25_data['active users']) & ~np.isnan(may24_to_may25_data['time_of_day'])
anova_result = f_oneway(*[group['active users'].values for name, group in may24_to_may25_data[mask].groupby('time_of_day')])
print(f"ANOVA result: F-statistic = {anova_result.statistic}, p-value = {anova_result.pvalue}")
# Interpret the ANOVA result
if anova_result.pvalue < 0.05:
    print("There is a significant difference in active users across different hours of the day.")
else:
    print("There is no significant difference in active users across different hours of the day.")

```

```python
# Plot the distribution of active users to see if it is normally distributed
may24_to_may25_data['active users'].hist(bins=30, edgecolor='black')
plt.xlabel('Active Users')  
plt.ylabel('Frequency')
plt.title('Distribution of Active Users')
plt.grid(axis='y')
plt.show()
```

```python
# Transform active users to make it more normally distributed
import numpy as np

may24_to_may25_data['active users log'] = may24_to_may25_data['active users'].apply(lambda x: 0 if x <= 0 else np.log(x))

# Plot the transformed distribution of active users
may24_to_may25_data['active users log'].hist(bins=30, edgecolor='black')
plt.xlabel('Active Users (Log Scale)')
plt.ylabel('Frequency')
plt.title('Distribution of Active Users (Log Scale)')
plt.grid(axis='y')
plt.show()
```

```python
# Test the normality of the transformed active users
from scipy import stats
_, p_value = stats.shapiro(may24_to_may25_data['active users log'].dropna())
print(f'Shapiro-Wilk test p-value: {p_value}')  
# Interpret the p-value
if p_value > 0.05:
    print("The transformed active users data is normally distributed (fail to reject H0).")
else:
    print("The transformed active users data is not normally distributed (reject H0).")
```

```python
# Now, test for correlation between time of day and active users
from scipy.stats import pearsonr
# Find index of rows with no active users
mask = (~np.isnan(may24_to_may25_data['time_of_day'])) & (~np.isnan(may24_to_may25_data['active users log']))
# Calculate Pearson correlation coefficient
correlation, p_value = pearsonr(may24_to_may25_data['time_of_day'][mask], may24_to_may25_data['active users log'][mask])
print(f'Pearson correlation coefficient: {correlation}, p-value: {p_value}')
# Interpret the correlation result
if p_value > 0.05:
    print("No significant correlation between time of day and active users (fail to reject H0).")
else:
    print("Significant correlation between time of day and active users (reject H0).")

```

```python
# Understand if it's possible to grasp which time of day has the most active users (using not log active users) Publication time
import matplotlib.pyplot as plt

# Group by time of day and calculate mean active users
median_active_users = may24_to_may25_data.groupby('time_of_day')['active users'].median()

median_active_users = median_active_users.sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(12, 6))
median_active_users.plot(kind='bar')
plt.title('Mean Active Users by Time of Day (Publication Time)')
plt.xlabel('Time of Day')
plt.ylabel('Mean Active Users')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Create a category assigning time of day to day bucket (create 7 or 8 buckets)
def time_of_day_bucket(hour):
    if 0 <= hour < 3:
        return 'Late Night'
    elif 3 <= hour < 6:
        return 'Early Morning'
    elif 6 <= hour < 9:
        return 'Morning'
    elif 9 <= hour < 12:
        return 'Late Morning'
    elif 12 <= hour < 15:
        return 'Afternoon'
    elif 15 <= hour < 18:
        return 'Late Afternoon'
    elif 18 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

may24_to_may25_data['time_of_day_bucket'] = may24_to_may25_data['time_of_day'].apply(time_of_day_bucket)

# Plot the distribution of active users by time of day bucket (publication time)
plt.figure(figsize=(12, 6))
may24_to_may25_data.groupby('time_of_day_bucket')['active users'].median().sort_values(ascending=False).plot(kind='bar')
plt.title('Median Active Users by Time of Day Bucket (Publication Time)')
plt.xlabel('Time of Day Bucket')
plt.ylabel('Median Active Users')
plt.xticks(rotation=45)
plt.show()
```
